```python
moviemat = df.pivot_table(index = "user_id" , columns = "title", values = "ratings")
```

```python
moviemat
```

```python
starwar_user_ratings = moviemat['Star Wars (1977)']
```

```python
starwar_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwar_user_ratings)
```

```python
corr_with_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
```

```python
corr_with_starwars.head()
```

```python
corr_with_starwars.dropna(inplace = True)
```

```python
corr_with_starwars[ corr_with_starwars['no of ratings']>100].sort_values('Correlation' , ascending = False)
```

```python
##Predict FUNCTION
```

```python
def predict_movies(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    
    corr_with_movie = pd.DataFrame(similar_to_movie, columns = ['Correlation'])
    corr_with_movie.dropna(inplace = True)
    
    corr_with_movie = corr_with_movie.join(ratings['no of ratings'])
    predictions = corr_with_movie[ corr_with_movie['no of ratings']>100].sort_values('Correlation' , ascending = False)
    
    return predictions
```

```python
predictions = predict_movies('Titanic (1997)')
```
